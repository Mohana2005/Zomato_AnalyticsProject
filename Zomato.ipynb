```python
df_country=pd.read_excel('Country-Code[1].xlsx')
df_country
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
country_names
#final_df.Country.value_counts().index --> gives country names only
```

```python
country_count=final_df.Country.value_counts().values
country_count
```

```python
plt.pie(country_count[:3],labels=country_names[:3],autopct='%.2f%%') # gives info about top three countries that use zomato
```

Observation : Zomato record maximum transactions are from India , followed by United States and then United Kingdom

```python
ratings=df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
ratings
```

#Observation
1. when rating is between 4.5 to 4.9 ---> Excellent
2. when raating is between 4.0 to 4.4 ---> Very Good
3. when rating is between 3.5 to 3.9 ---> Good
4. when rating is between 2.5 to 3.4 ---> Average
5. when rating is between 1.8 to 2.4 ---> Poor
6. when rating is 0 ---> It is not rated

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating_count',data=ratings,hue='Rating color',palette=['grey','Red','Orange','Yellow','Green','blue'])
```

#observation
1. Not rated count is very high
2. maximum count is between ratings 2.5 to 3.5


```python
sns.countplot(x='Rating color',data=ratings)
```

This countplot doesnt give the actual count of the colors , instead reprsesnts its frequecies as per the ratings table
